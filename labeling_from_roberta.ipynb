```python
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

##MODEL 1 (2 CLASSES: 1, 0)
84.14% on dev/validation and 84.64% on test.

https://huggingface.co/w11wo/indonesian-roberta-base-indolem-sentiment-classifier-fold-0

```python
#model 1

pretrained_name = "w11wo/indonesian-roberta-base-indolem-sentiment-classifier-fold-0"

nlp = pipeline(
    "sentiment-analysis",
    model=pretrained_name,
    tokenizer=pretrained_name)

def labeling(df):
  label = nlp(df.cleantext)
  values = list(label[0].values())
  return values
df2['label2'] = df2.apply(labeling, axis=1)
df2[['newlabel1','score1']] = df2.label2.tolist()
del df2['label2']
```

## MODEL 2 (3 CLASSES: pos, neg, net)

https://huggingface.co/ayameRushia/bert-base-indonesian-1.5G-sentiment-analysis-smsa

```python
#model 2

tokenizer2 = AutoTokenizer.from_pretrained("ayameRushia/bert-base-indonesian-1.5G-sentiment-analysis-smsa")
model2 = AutoModelForSequenceClassification.from_pretrained("ayameRushia/bert-base-indonesian-1.5G-sentiment-analysis-smsa")

nlp2 = pipeline(
    "sentiment-analysis",
    model=model2,
    tokenizer=tokenizer2)

def labeling2(df):
  label = nlp2(df.cleantext)
  values = list(label[0].values())
  return values
df2['label'] = df2.apply(labeling2, axis=1)
df2[['newlabel2','score2']] = df2.label.tolist()
del df2['label']
```

## MODEL 3 (3 CLASSES: pos, neg, net)

https://huggingface.co/akahana/indonesia-sentiment-roberta



```python
#model 3, tried num_labels=2

tokenizer3 = AutoTokenizer.from_pretrained("akahana/indonesia-sentiment-roberta")
model3 = AutoModelForSequenceClassification.from_pretrained("akahana/indonesia-sentiment-roberta", num_labels=2, ignore_mismatched_sizes=True)

nlp3 = pipeline(
    "sentiment-analysis",
    model=model3,
    tokenizer=tokenizer3)

def labeling3(df):
  label = nlp3(df.cleantext)
  values = list(label[0].values())
  return values

df2['label'] = df2.apply(labeling3, axis=1)
df2[['newlabel3','score3']] = df2.label.tolist()
del df2['label']
```

```python
df2['newlabel1'] = df2['newlabel1'].replace(['LABEL_0'], '-1')
df2['newlabel1'] = df2['newlabel1'].replace(['LABEL_1'], '1')
```
