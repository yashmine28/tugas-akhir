```python
import pandas as pd
df1=pd.read_csv('/content/random preprocessed dataset.csv', sep=';', usecols=['cleantext','tokens','label'])
#df= pd.read_csv('/content/drive/MyDrive/college/TTTTTTAAAAAAAA/dataset TA/labeled 5000.csv', sep=';', usecols=['cleantext','tokens'])
#df1 = pd.read_csv('/content/drive/MyDrive/college/TTTTTTAAAAAAAA/dataset TA/preprocessed 6142.csv', sep=';', usecols=['cleantext','tokens'])
pos_dict = pd.read_csv('/content/drive/MyDrive/college/TTTTTTAAAAAAAA/dataset TA/positive21.csv', sep=';', usecols=['stemdict','weight'])
neg_dict = pd.read_csv('/content/drive/MyDrive/college/TTTTTTAAAAAAAA/dataset TA/negative21.csv', sep=';', usecols=['stemdict','weight'])
new_dict = pd.read_csv('/content/drive/MyDrive/college/TTTTTTAAAAAAAA/dataset TA/unlisted_terms.csv', sep=';')
```

```python
df1.head()
```

```python
import ast

def converter(df):
  token = ast.literal_eval(df)
  return [i for i in token]

df1['tokens'] = df1['tokens'].apply(converter)
df1.head()
```

```python
import nltk 
from nltk import word_tokenize
from nltk.util import ngrams
def bigram(df):
  return list(ngrams(df.tokens, 2))
df1['tokens2'] = df1.apply(bigram, axis=1)
df1.head()
```

```python
def biword(df):
  alist = []
  for bigram in df.tokens2:
    x = "%s %s" % bigram
    alist.append(x)
  return alist

df1['tokens22'] = df1.apply(biword, axis=1)
df1.head()
```

```python
for a in df1:
  tokens12 = df1.tokens + df1.tokens22
df1 = df1.merge(tokens12.to_frame('tokens12'), left_index=True, right_index=True)
```

```python
df1.tokens12[1]
```

```python
neg = dict(zip(neg_dict['stemdict'],neg_dict['weight']))
pos = dict(zip(pos_dict['stemdict'],pos_dict['weight']))
new = dict(new_dict.values)
```

```python
#trying to make a unlisted terms
posneg = {**pos,**neg}
listlib = list(posneg.keys())
```

```python
#del df1['label']
#del df1['score']

def polarity(df):
  tweet_score = 0
  
  for term in df.tokens12:
    for word, weight in pos.items():
      if term == word:
        tweet_score += weight

  for term in df.tokens12:
    for word, weight in neg.items():
      if term == word:
        tweet_score += weight

  return int(tweet_score)

def label(df):
    if df.score >= 0:
      return int(1)
    else:
      return int(-1)
    

df1['score'] = df1.apply(polarity, axis=1)
df1['label'] = df1.apply(label, axis=1)
```

```python
#FOR UNLISTED-WORD MAKING ONLY. make all tokens to dataframe
all_tokens = df1.tokens.explode()
allwordlist = all_tokens.value_counts().reset_index()
allwordlist.columns = ['Word', 'Frequency']
allwordlist
```
